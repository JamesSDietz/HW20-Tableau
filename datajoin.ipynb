```python
df201901 = pd.read_csv("201901-citibike-tripdata.csv")
df201901.tail()
```

```python
df = pd.concat([df201901, df201902, df201903, df201904, df201905, df201906])
df = df.sort_values(by='starttime').reset_index(drop=True)
df.tail()
```

```python
df = df.rename(columns = {'birth year':'birth_year'})
```

```python
df['age'] = 2019 - df['birth_year']
```

```python
df.head()
```

```python
df['gender'].replace(0, 'unknown',inplace=True)
df['gender'].replace(1, 'Male',inplace=True)
df['gender'].replace(2, 'Female',inplace=True)
```

```python
df.head()
```

```python
df['tripduration_mins'] = df['tripduration']/60
df.head()
```

```python
df['trip_mileage'] = df['tripduration_mins'] * 0.12426667
df
```

```python
df.dtypes
```

```python
df.to_csv('citibikedata2019.csv')
```
